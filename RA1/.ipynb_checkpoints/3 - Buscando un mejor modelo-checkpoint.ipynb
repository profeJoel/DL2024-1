# Parte 1 - Redes Neuronales para Regresión

```python
from tensorflow.keras.datasets import boston_housing

(X_train, y_train), (X_test, y_test) = boston_housing.load_data()
```

```python
X_train.shape
```

```python
X_train
```

```python
import pandas as pd

X_train_df = pd.DataFrame(X_train)
```

```python
X_train_df.head()
```

```python
#Normalizar los datos

promedio = X_train.mean(axis=0)
X_train -= promedio

DS = X_train.std(axis=0)
X_train /= DS


X_test-=promedio
X_test/=DS
```

```python
from tensorflow import keras

# Crear un modelo de Redes Neuronales para Regresión
def crear_modelo():
    modelo = keras.Sequential([
        keras.layers.Dense(64, activation="relu"),
        keras.layers.Dense(64, activation="relu"),
        keras.layers.Dense(1),
    ])
    modelo.compile(optimizer="rmsprop", loss="mse", metrics=["mae"])
    return modelo
```

```python
modelo_regresion = crear_modelo()
```

```python
modelo_regresion.summary()
```

```python
#entrenamiento
entrenamiento = modelo_regresion.fit(X_train, y_train, epochs=1000, batch_size = 10, validation_data=(X_test, y_test))
```

```python
import matplotlib.pyplot as plt

# Grafica de entrenamiento
# Graficar la pérdida de la evolución del entrenamiento
plt.plot(entrenamiento.history["mae"], 'r', marker='.', label = 'Train MAE')
plt.plot(entrenamiento.history["val_mae"], 'b', marker='.', label = 'Validation MAE')
plt.show();
```

```python
import numpy as np

# Entrenamiento con k-fold validation
k = 4
tajada_muestras_validacion = len(X_train)//k

epocas = 1000
metricas = []

#kfold validation es un for
# Es hacer k entrenamientos con distintos conjuntos de entrenamiento y validación
for i in range(k):
    print(f"> Procesando k fold N{i}")
    X_val = X_train[i * tajada_muestras_validacion: (i+1) * tajada_muestras_validacion]
    y_val = y_train[i * tajada_muestras_validacion: (i+1) * tajada_muestras_validacion]
    
    X_train_parcial = np.concatenate(
        [X_train[:i * tajada_muestras_validacion],
        X_train[(i+1) *tajada_muestras_validacion:]],
        axis=0)
    y_train_parcial = np.concatenate(
        [y_train[:i * tajada_muestras_validacion],
        y_train[(i+1) *tajada_muestras_validacion:]],
        axis=0)
    
    regresion_kfold = crear_modelo()
    regresion_kfold.fit(X_train_parcial, y_train_parcial, epochs=epocas, batch_size=16)
    val_mse, val_mae = regresion_kfold.evaluate(X_val, y_val)
    metricas.append(val_mae)
    
```

```python
# Entrenamiento con k-fold validation
k = 4
tajada_muestras_validacion = len(X_train)//k

epocas = 1000
metricas = []

#kfold validation es un for
# Es hacer k entrenamientos con distintos conjuntos de entrenamiento y validación
for i in range(k):
    print(f"> Procesando k fold N{i}")
    X_val = X_train[i * tajada_muestras_validacion: (i+1) * tajada_muestras_validacion]
    y_val = y_train[i * tajada_muestras_validacion: (i+1) * tajada_muestras_validacion]
    
    X_train_parcial = np.concatenate(
        [X_train[:i * tajada_muestras_validacion],
        X_train[(i+1) *tajada_muestras_validacion:]],
        axis=0)
    y_train_parcial = np.concatenate(
        [y_train[:i * tajada_muestras_validacion],
        y_train[(i+1) *tajada_muestras_validacion:]],
        axis=0)
    
    regresion_kfold = crear_modelo()
    entrenamiento = regresion_kfold.fit(X_train_parcial, y_train_parcial, epochs=epocas, batch_size=16, validation_data = (X_val, y_val))
    historial_mae = entrenamiento.history["val_mae"]
    metricas.append(historial_mae)
    
```

```python
promedio_mae_historico = [np.mean([x[i] for x in metricas]) for i in range(epocas)]
```

```python
plt.plot(entrenamiento.history["mae"], 'r', marker='.', label = 'Train MAE')
plt.plot(range(1, len(promedio_mae_historico)+1), promedio_mae_historico)
plt.xlabel("Epocas")
plt.ylabel("MAE val")
plt.show
```

```python
val_mae_truncado = promedio_mae_historico[10:]
entrenamiento_truncado = entrenamiento.history["mae"][10:]
plt.plot(entrenamiento_truncado, 'r', marker='.', label = 'Train MAE')
plt.plot(range(1, len(val_mae_truncado)+1), val_mae_truncado)
plt.xlabel("Epocas")
plt.ylabel("MAE val")
plt.show
```

```python
from tensorflow.keras import regularizers

# Crear un modelo de Redes Neuronales para Regresión
def crear_modelo_regularizado():
    modelo = keras.Sequential([
        keras.layers.Dense(64, activation="relu", kernel_regularizer = regularizers.l2(0.002)),
        keras.layers.Dense(64, activation="relu", kernel_regularizer = regularizers.l2(0.002)),
        keras.layers.Dense(1),
    ])
    modelo.compile(optimizer="rmsprop", loss="mse", metrics=["mae"])
    return modelo
```

```python
# regularizado con l1 a 0.001
regresion_regularizada = crear_modelo_regularizado()
entrenamiento = regresion_regularizada.fit(X_train, y_train, epochs=300, batch_size=16, validation_data = (X_test, y_test))
plt.plot(entrenamiento.history["mae"], 'r', marker='.', label = 'Train MAE')
plt.plot(entrenamiento.history["val_mae"], 'b', marker='.', label = 'Validation MAE')
plt.show();
```

```python
# regularizado con l2 a 0.001
regresion_regularizada = crear_modelo_regularizado()
entrenamiento = regresion_regularizada.fit(X_train, y_train, epochs=300, batch_size=16, validation_data = (X_test, y_test))
plt.plot(entrenamiento.history["mae"], 'r', marker='.', label = 'Train MAE')
plt.plot(entrenamiento.history["val_mae"], 'b', marker='.', label = 'Validation MAE')
plt.show();
```

```python
# Modelo Dropout
# Crear un modelo de Redes Neuronales para Regresión
def crear_modelo_dropout():
    modelo = keras.Sequential([
        keras.layers.Dense(64, activation="relu", kernel_regularizer = regularizers.l2(0.002)),
        keras.layers.Dropout(0,5),
        keras.layers.Dense(64, activation="relu", kernel_regularizer = regularizers.l2(0.002)),
        keras.layers.Dropout(0,5),
        
        keras.layers.Dense(1),
    ])
    modelo.compile(optimizer="rmsprop", loss="mse", metrics=["mae"])
    return modelo
```

```python
# modelo con dropout
regresion_dropout = crear_modelo_dropout()
entrenamiento = regresion_dropout.fit(X_train, y_train, epochs=300, batch_size=16, validation_data = (X_test, y_test))
plt.plot(entrenamiento.history["mae"], 'r', marker='.', label = 'Train MAE')
plt.plot(entrenamiento.history["val_mae"], 'b', marker='.', label = 'Validation MAE')
plt.show();
```
